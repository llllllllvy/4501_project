```python
#？？要用吗 我没用懒得用了
def mixo(x):
    if type(x)==str:
        x = x.replace('O','0')
    return x
```

filter the row which there is letter in zipcodes

```python
import string
all_letters = string.ascii_letters
special = string.punctuation
#判断是否含有字母，如有返回True
def letterin(x):
    if type(x)==str and len(set(x)&set(all_letters))>0:
        return True
    else:
        return False
    
#在不足五位的zip前填充0
def fill0(x):
    return '0'*(5-len(x))+x

#判断是否有数字外的其他符号，如有返回False
def dele(x):
    if len(set(x)-set('1234567890'))>0:
        return False
    else:
        return True
```

```python
#去除含有字母的zip
contains_letter = data311df_remove1['incident_zip'].apply(letterin)
rows_with_letter = data311df_remove1[contains_letter]
data311df_remove2 = data311df_remove1.drop(rows_with_letter.index)

#只保留空格.-前的数字 ; O -> 0
col1 = data311df_remove2.incident_zip.apply(str).apply(mixo).apply(lambda x:x.split(' ')[0]).apply(lambda x:x.split('.')[0]).apply(lambda x:x.split('-')[0]).apply(fill0)
#去除zip前的`
col2 = col1.apply(lambda x:x.strip('`'))
#删除有数字外的其他符号zip的所在行
col3 = col2[col2.apply(dele)]

data311df_remove3 = data311df_remove2.loc[col3.index]
data311df_remove3.loc[col3.index,'incident_zip'] = col3
```

## Query 1

```python
query1 = """
        SELECT incident_zip as zip_code,
               COUNT(unique_key) AS complaint_num
        FROM tb311
        WHERE created_date BETWEEN '2022-10-01T00:00:00.000' AND '2023-09-30T23:59:59.999'
        GROUP BY incident_zip
        """
```

```python
query2 = """
        SELECT zipcode,
               COUNT(tree_id) AS tree_num
        FROM tbTree
        GROUP BY zipcode
        ORDER BY COUNT(tree_id) DESC
        LIMIT 10
        """
```

```python
query3 = """
        WITH tb_most_tree_zipcode as(
                                    SELECT zipcode,
                                           COUNT(tree_id) AS tree_num
                                    FROM tbTree
                                    GROUP BY zipcode
                                    ORDER BY COUNT(tree_id) DESC
                                    LIMIT 10),
        SELECT t1.zipcode,
               ROUND(AVERAGE(t2.2023-08-31),2) AS average_rent
        FROM tb_most_tree_zipcode AS t1
        JOIN tbRent AS t2 ON t1.zipcode = t2.RegionName
        GROUP BY t1.zipcode
        ORDER BY t1.tree_num
        """
        
```

```python
query4 = """
        WITH 
        lowest_highest_rent_zipcode as(SELECT RegionName as zipcode,
                                              ROUND(AVERAGE(2023-01-31),2) AS average_rent
                                       FROM tbRent
                                       GROUP BY zipcode
                                       ORDER BY ROUND(AVERAGE(2023-01-31),2) ASC
                                       LIMIT 5
                                       UNION ALL
                                       SELECT RegionName as zipcode,
                                              ROUND(AVERAGE(2023-01-31),2) AS average_rent
                                       FROM tbRent
                                       GROUP BY zipcode
                                       ORDER BY ROUND(AVERAGE(2023-01-31),2) DESC
                                       LIMIT 5),
                                               
        SELECT t1.zipcode,
               t1.average_rent,
               COUNT(t2.tree_id) AS tree_num,
               COUNT(t3.unique_key) AS complaint_num
        FROM lowest_highest_rent_zipcode as t1
        LEFT JOIN tbTree AS t2 ON t1.zipcode = t2.zipcode
        LEFT JOIN tb311 AS t3 ON t1.zipcode = t2.incident_zip
        GROUP BY t1.zipcode, t1.average_rent
        ORDER BY t1.average_rent
        """
               
                                           
```

```python
query5 = """
        SELECT t1.ZIPCODE,
               COUNT(t2.tree_id) AS tree_num
        FROM tbZipcodes AS t1
        JOIN tbTree AS t2 ON ST_Within(ST_MakePoint(t2.longitude, t2.Latitude), t1.geometry)
        GROUP BY t1.ZIPCODE
        """
```

```python
query6 = """
        SELECT tree_id AS ID,
               spc_latin AS species,
               health,
               status,
               (longitude,Latitude) AS location
        FROM tbTree
        WHERE ST_DWithin(
            ST_MakePoint(longitude, Latitude)::geography,
            ST_MakePoint(-73.96253174434912, 40.80737875669467)::geography,
            805  
         """
```
